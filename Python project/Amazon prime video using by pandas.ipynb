#### 08. Calculate the average usage frequency of all users.

```python
df.usage_frequency.value_counts()
```

```python
# Define a mapping for usage frequency categories to numerical values

usage_frequency_mapping = {
    'Frequent' : 3,
    'Regular' : 2,
    'Occasinal': 1
}

# Create a new column with numerical values based on the mapping 

df['usage_frequency_numeric'] = df['usage_frequency'].map(usage_frequency_mapping)


# Calculate the average usgae frequency 

average_frequency_usgae =round(df['usage_frequency_numeric'].mean(),2)

print('Usage Frequency' , + average_frequency_usgae)
```

#### 15. List the users who have used a specific device (Tablet).

```python
# Group by 'username' and aggregate 'devices_used' into a list if there are multiple unique values 

grouped = df.groupby('username')['devices_used'].agg(lambda x: list(set(x)) if len(x.unique()) > 1 else x.iloc[0]).reset_index() 

# Calculate the count of devices used and sort by the count in descending order

grouped['devices_count'] = grouped['devices_used'].apply(lambda  x: len(x) if isinstance(x, list) else 1)


result = grouped.sort_values(by='devices_count', ascending=False).reset_index(drop = True)

result
```
