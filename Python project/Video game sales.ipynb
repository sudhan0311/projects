# Dataset Description
The dataset consists of the following fields:

Rank: The ranking of overall sales for each game.

Name: The name of the video game.

Platform: The platform on which the game was released (e.g., PC, PS4, Xbox One, etc.).

Year: The year of the game's release.

Genre: The genre of the video game (e.g., Action, Sports, RPG, etc.).

Publisher: The publisher of the video game.

NA_Sales: Sales in North America (in millions).

EU_Sales: Sales in Europe (in millions).

JP_Sales: Sales in Japan (in millions).

Other_Sales: Sales in the rest of the world (in millions).

Global_Sales: Total worldwide sales (sum of all regional sales).

The dataset contains 16,598 records, with 2 records dropped due to incomplete information.

# Let's Start!

**1. Top Selling Games: What are the top 10 best-selling games of all time based on global sales?**

```python
# Sort the dataset by 'Global_Sales' in descending order and select the top 10

top_10_games = df.sort_values(by='Global_Sales', ascending=False).head(10)

# Create a bar chart to visualize the  global sales of top 10 games

pp.figure(figsize=(14, 6))
pp.bar(top_10_games["Name"],top_10_games['Global_Sales'], color='orange')
pp.xlabel("Game Name")
pp.ylabel('Global Sales (millions)')
pp.title("Top 10 Best-Selling Games of All Time")
pp.xticks(rotation=90)
pp.show()
```

2. **Sales by Platform: Which gaming platform (e.g., PC, PS4, Xbox, etc.) has the highest total sales?**

```python
# Group the databy 'platform' and calculate the sum of global sales for the each platform

platform_sales = df.groupby('Platform')['Global_Sales'].sum().reset_index()

# sort the platforms based on total sales in descending order

platform_sales = platform_sales.sort_values(by='Global_Sales', ascending=False)

# Create bar short to visualize total sales by platform

pp.figure(figsize=(14, 6))
pp.bar(platform_sales['Platform'],platform_sales['Global_Sales'], color='RED')
pp.title(" TOtal Sales by Gaming Platform")
pp.xlabel("Gaming Platform")
pp.ylabel("Total Sales (million)")
pp.xticks(rotation=45)
pp.show()
```

3. **Genre Popularity: What are the most popular genres of video games in terms of global sales?**

```python
# Group the data by 'Genre' annd calculate the sum of the global sales for each genre

genre_sales = df.groupby( 'Genre')["Global_Sales"].sum().reset_index()

# sort the genres based on the total salees in descending order 

genre_sales = genre_sales.sort_values(by='Global_Sales', ascending=False)

#  Create a pie chart to visualize genre popularity

pp.figure(figsize=(14, 6))
pp.pie(genre_sales['Global_Sales'], labels=genre_sales['Genre'],autopct='%1.1f%%', startangle=10,colors=pp.cm.Paired.colors)
pp.title(" Genre popularity based  on Global Sales")
pp.axis("equal")
pp.show()
```

4. **Sales by Region: Which region (North America, Europe, Japan, Rest of the World) contributes the most to global video game sales?**

```python
# Calculate each total sales for each region

region_sales = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()

# Create a stacked bar chart to visualize sales by region 

pp.figure(figsize=(14, 6))
pp.bar(region_sales.index, region_sales.values, color=['royalblue', 'mediumseagreen','tomato','gold'])
pp.xlabel("Region")
pp.ylabel('Total Sales (Millions)')
pp.title("Total Sales by Region")
pp.xticks(rotation=20)
pp.ylim(0,8000)
pp.show()
```

5. **Publisher Performance: Which publishers have released the most successful games in terms of global sales?**

```python
# Group the data by 'publisher' and calculate the sum of Global sales for each publisher

publisher_sales = df.groupby('Publisher')["Global_Sales"].sum().reset_index()

# Sort the publisher based on the  total sales in descending order
publisher_sales = publisher_sales.sort_values(by='Global_Sales', ascending=False).head(10)

# Create a Horizontal bar chart to visualize puplisher performance

pp.figure(figsize=(14, 6)) 
pp.barh(publisher_sales['Publisher'],publisher_sales['Global_Sales'], color="Black")
pp.xlabel("Total Sales (millions)")
pp.ylabel("publisher")
pp.xlim(0,3000)
pp.title("Top Publisher by Global Sales")
pp.show()
```

6. **Yearly Sales Trends: How have video game sales evolved over the years? Are they increasing or decreasing?**

```python
# Group the data by ' Year ' and calculate the sum of global sales for each year

yearly_sales = df.groupby('Year')["Global_Sales"].sum()

# Create a line line chart to visualize the yearly sales trends 
pp.figure(figsize=(14, 6))
pp.plot(yearly_sales.index, yearly_sales.values, 'o-r')
pp.xlabel("Year")
pp.ylabel(" Total Sales ( Millions)")
pp.ylim(0,800)
pp.title("Yearly Video Game Sales Trends")
pp.show()
```

7. **Platform vs. Genre: Which genres are most popular on specific gaming platforms?**

```python
# Group the data by 'platform' and 'Gener' and calculate the sum of Global sales for each compination 

platform_genre_sales  =  df.groupby(['Platform','Genre'])['Global_Sales'].sum().reset_index()

# Create a grouped bar chart to visualize genre popularity on gaming platforms
pp.figure(figsize=(14, 6))
sb.barplot(x='Platform', y='Global_Sales', hue="Genre",data=platform_genre_sales, palette='tab20')
pp.xlabel("Platform")
pp.ylabel("Total Sales (millions)")
pp.title('Genre popularity on Gaming Platform')
pp.legend(title="Genre", bbox_to_anchor=(1.010 , 1), loc="upper left")
pp.xticks(rotation=45)
pp.show()
```

8 .**Regional Sales Variation: Do certain genres or platforms perform better in specific regions?**

```python
# Create a Box plot to visualize sales variation of genre and platforms in specific region

pp.figure(figsize=(14, 6))
sb.boxplot(x="Genre", y='NA_Sales',data=df)
pp.xlabel('Genre')
pp.ylabel(" Sales in North America (Millions)")
pp.title("Regional Sales Variation by Genre in North America")
pp.ylim(0,50)
pp.xticks(rotation= 45)
pp.show()
```

9. **Correlations: Are there any significant correlations between game sales and other factors like year of release, genre, or publisher?**

```python
# Create a pairplot to visualize correlation between game sales, year of release , and publisher 

sb.pairplot(df[["Global_Sales",'Year','Publisher']])
pp.show()
```

10. **Distribution of Sales: What is the distribution of global sales across different games? Are sales concentrated in a few blockbuster titles, or is it more evenly spread?**

```python
# Create a Histogram to visualize the distribution of global sales 

pp.figure(figsize=(14, 6))
pp.hist(df['Global_Sales'],bins=10, color="orange" , edgecolor="black")
pp.xlabel(" Global Sales (millions)")
pp.ylabel('Frequency')
pp.title("Distribution of Global Sales")
pp.show()
```

11. **Sales Over Time: Can you identify any trends in sales for specific games or genres over time?**

```python
# Group the Data by 'Year' and calculate the sum of Global sales for each year

yearly_sales = df.groupby('Year')['Global_Sales'].sum()

# Create line plot to visualize the sales over time 
pp.figure(figsize=(14, 6))
pp.plot(yearly_sales.index, yearly_sales.values, "o-r")
pp.xlabel("Year")
pp.ylabel('Total sales (Millions)')
pp.title(" Video Game Sales Over Time")
pp.show()
```

12. **Best-Selling Genres by Region: Are there any differences in the most popular genres for each region (NA, EU, JP)?**

```python
# Group the data by 'Genre' and calculate the sum of global sales for each genre
genre_sales=df.groupby('Genre')['Global_Sales'].sum().reset_index()

# sort the genres based on total sales in descending order
genre_sales=genre_sales.sort_values(by='Global_Sales',ascending=False)

# Create a bar chart to visualize the best-selling genres globally
pp.figure(figsize=(14, 6))
pp.bar(genre_sales['Genre'],genre_sales['Global_Sales'], color='navy')
pp.xlabel("Genre", color = 'navy')
pp.ylabel("Total Sales (million)", color = 'Navy')
pp.title("Best-Selling Genres Globally ", color="black")
pp.xticks(rotation=45)
pp.show()
```

13. **Yearly Sales by Region: How have the sales trends evolved over the years in each region?**

```python
# Group the data by "Year" and caluculate the sum of sales for each region for every year
yearly_sales_by_region= df.groupby('Year')[['NA_Sales', 'EU_Sales' , 'JP_Sales' , 'Other_Sales']].sum()

# Create a line plot to visualize yearly sales trends in each region
pp.figure(figsize=(14, 6))
pp.plot(yearly_sales_by_region.index, yearly_sales_by_region['NA_Sales'], label='North America', marker='o' , color="Royalblue" )
pp.plot(yearly_sales_by_region.index, yearly_sales_by_region['EU_Sales'], label='Europe', marker='o', color='mediumseagreen' )
pp.plot(yearly_sales_by_region.index, yearly_sales_by_region['JP_Sales'], label='Japan', marker='o', color='yellow' )
pp.plot(yearly_sales_by_region.index, yearly_sales_by_region['Other_Sales'], label='Other Sales', marker='o', color='red')
pp.xlabel('Year')
pp.ylabel('Sales (million)')
pp.title(" Yearly Sales Trends by Region")
pp.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
pp.grid(True)
pp.show()
```

14. **Regional Market Share: What percentage of the global video game market do each of the regions hold?**


```python
# Calculate the total global sales 

total_global_sales= df["Global_Sales"].sum()

# Calculate the market share of each region 
region_market_share=df[['NA_Sales', 'EU_Sales', 'JP_Sales' , 'Other_Sales']].sum() / total_global_sales*100

# Create  a pie  chart to visualize the regional market share
pp.figure(figsize=(14, 6))
pp.pie(region_market_share, labels=region_market_share.index, autopct= '%1.1f%%' , colors=pp.cm.Paired.colors)
pp.axis('equal')
pp.title("Regional Market share of Video Game Sales")
pp.show()
```

15. **Platform Sales Comparison: Compare the sales performance of different platforms side by side.**

```python
# Group the data by 'Platform' and calculate the sum of global sales for each platform
platform_sales=df.groupby('Platform')['Global_Sales'].sum().reset_index()

#Sort the  platforms based on total sales in descending order 
platform_sales=platform_sales.sort_values(by="Global_Sales", ascending=False)

# Create  a Grouped bar chart  to compare platform sales
pp.figure(figsize=(14, 6))
pp.bar(platform_sales['Platform'],platform_sales['Global_Sales'],color='navy')
pp.xlabel("Platform")
pp.ylabel('Total Sales (Millions)')
pp.title("platform Sales Comparison")
pp.xticks(rotation=45)
pp.show()
```
